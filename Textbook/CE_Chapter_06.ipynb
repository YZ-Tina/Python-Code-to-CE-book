### Chapter 6. The overlapping generations model

Reference: 
Fehr, H. and Kindermann, F. 2018. The overlapping generations model. <i>Introduction to Computational Economics Using Fortran.</i> Oxford University Press. pp225-252.

This chapter focus on the individual-choice problem without taking into account the interaction between households, the production sector of the economy, and the government. In this chapter we take a broader perspective and embed the life-cycle model into a general equilibrium framework. In this framework, prices adjust in order to balance supply and demand in goods and factor markets and the government has to operate under some balance budget rules. Indivual savings behaviour endogenously determines the capital stock. We introduce the most basiv version of the OLG model and discuss the computation of a transition path and the intergenerational welfare effects of policy reofrms.

## 6.1 General structure and long-run equilibrium

### 6.1.1 Demographics, Behaviour and Markets

<b>Demographics</b> We assume that households in the model live for three periods. For simplicity we do not account for income and lifespan uncertainty. In each successive period $t$ a new cohort is born, where the number of households $N_t$ in this cohort grows at a rate $n_{p,t}$, i.e. $$N_t=(1+n_{p,t})N_{t-1}.$$ 

<b>Households' decisions</b> Households decide about how much to consume and save in the different periods of their life. They have to pay taxes on income and consumption as well as a payroll tax to the pension system, in reward for which they receive pension benefits $pen_{t+2}$ in the third period of life. The <i>dynamic budget constraints</i> in the three different periods are then given by \begin{equation}
\begin{split}
p_tc_{1,t}&=w_t^n-a_{2,t+1} \\
p_{t+1}c_{2,t+1}&=R_{t+1}^na_{2,t+1}+w_{t+1}^n-a_{3,t+2} \\
p_{t+2}c_{3,t+2}&=R_{t+2}^na_{3,t+2}+pen_{t+2}
\end{split}
\end{equation}
where $w_t^n=w_t(1-\tau_t^w-\tau_t^p)$, $R_t^n=1+r_t(1-\tau_t^r)$, and $p_t=1+\tau_t^c$ define net wage rates, net interest rates and consumer prices, respectively. $\tau_t^c, \tau_t^w, \tau_t^p$, and $\tau_t^r$ denote consumption, labour income, payroll, and capital income tax rates. As in Chapter 5, households born in period ^t^ maximize the utility function $$U_t=U(c_{1,t},c_{2,t+1}, c_{3,t+2})=u(c_{1,t})+\beta u(c_{2,t+1})+\beta^2u(c_{3,t+2})$$ subject to the <i>intertemporal budget constraint</i> $$p_tc_{1,t}+\frac{p_{t+1}c_{2,t+1}}{R_{t+1}^n}+\frac{p_{t+2}c_{3,t+2}}{R_{t+1}^nR_{t+2}^n}=w_t^n+\frac{w_{t+1}^n}{R_{t+1}^n}+\frac{pen_{t+2}}{R_{t+1}^nR_{t+2}^n}=:W_{1,t},$$
which can be derived from the dynamic budget constraints. The resulting first-order conditions to be \begin{equation}
\begin{split}
p_{t+1}u'(c_{1,t})&=\beta R_{t+1}^np_tu'(c_{2,t+1}) \\
p_{t+2}u'(c_{2,t+1})&=\beta R_{t+2}^np_{t+1}u'(c_{3,t+1}).
\end{split}
\end{equation}

<b>Aggregation</b> Since different households are living in the same period $t$, the demand for goods and the supply of labour and capital need to be determined by aggregating individual household decisions. We normalize all aggregate variables to the size of the newborn cohort. Since each working household is assumed to supply one unit of labour inelastically, aggregate labour supply in period $t$ can be computed from $$L_t=\frac{1}{N_t}(N_t+N_{t-1})=1+\frac{1}{1+n_{p,t}}=\frac{2+n_{p,t}}{1+n_{p,t}}.$$ $$A_t=\frac{1}{N_t} (N_{t-1}a_{2,t}+N_{t-2}a_{3,t})=\frac{a_{2,t}}{1+n_{p,t}}+\frac{a_{3.t}}{(1+n_{p,t})(1+n_{p,t-1})},$$ where $a_{2,t}$ and $a_{3,t}$ denote savings of the two older cohorts built in the previous year, and per capita consumption from $$C_t=\frac{1}{N_t}(N_tc_{1,t}+N_{t-1}c_{2,t}+N_{t-2}c_{3,t})=c_{1,t}+\frac{c_{2,t}}{1+n_{p,t}}+\frac{c_{3,t}}{(1+n_{p,t})(1+n_{p,t-1})}.$$

<b>Firm's decisions</b> The production technology is represented by the neoclassical production function $F(K_t,L_t)$, which is homogeneous of degree one. The capital stock depreciates physically during the production process at rate $\delta$. Consequently, a firm's output $Y_t$ can be derived from $$Y_t=F(K_t,L_t) \quad \text{or} \quad y_t=\frac{Y_t}{L_t}=f(k_t)$$ where $y_t$ denotes per capita output and $k_t=\frac{K_t}{L_t}$ capital intensity. They produce under perfect competition conditions. Therefore equilibrium factor prices $r_t$ and $w_t$ are derived from first-order and zero-profit conditions are \begin{equation}
\begin{split}
r_t&=f'(k_t)-\delta \\
w_t&=f(k_t)-f'(k_t)k_t.
\end{split}
\end{equation}

<b>The government</b> The government sector comprises two separate budgets. It finances the provision of a public good $G_t$ and interest payments on public debt $r_tB_t$ by means of tax revenues and deficit spending, i.e. $$T_t+(1+n_{p,t+1})B_{t+1}-B_t=G_t+r_tB_t, $$ where $T_t=\tau_t^cC_t+\tau_t^ww_tL_t+\tau_t^rr_tA_t$ defines tax revenues from consumption and income taxation. The amount of public good is computed depending on the population structure, i.e. $$G_t=\frac{1}{N_t}(N_tg_1+N_{t_1}g_2+N_{t-2}g_3)=g_{1}+\frac{g_{2}}{1+n_{p,t}}+\frac{g_{3}}{(1+n_{p,t})(1+n_{p,t-1})}$$ where $g_j$ represent per capita coefficients of cohort $i$ for public good provision. Further more, we determine the deficit path exogenously such that $B_{t+1}=b_{y,t}Y_{t+1}$ holds in every period. We assume that in each period $t$ the level of pension benefits is computed as a fraction $\kappa_t$ of the previous period's gross wage $pen_t=\kappa_tw_{t-1}$. The pension system is fully pay-as-you-go financed. Hence, contributions from workers have to finance benefit payments to retirees $$\tau_t^pw_t(N_t+N_{t-1})=pen_tN_{t-2} \quad \text{or} \quad \tau_t^pw_t=\frac{pen_t}{(2+n_{p,t})(1+n_{p,t-1})}.$$

<b>Market equilibrium</b> For a market equilibrium the goods market and factor markets need to be balanced. On the capital market of a closed economy, interest rates are determined so as to equalize households' aggregate savings $A_t$ with the demand for capital of the firms $K_t$ and the government $B_t$, i.e. $$A_t=K_t+B_t.$$ In the labour market, wages are set to guarantees the aggregate labour supply of households matches labour demand of the firms. Finally, in the goods market total output of the firms must be used either for privare and public consumption or for building the new capital stock, i.e. $$Y_t=C_t+G_t+(1+n_{p,t+1})K_{t+1}-(1-\delta)K_t.$$

### 6.1.2 Computation of the Long-Run Equilibrium

<b>Functional forms</b> We define concrete functional forms of preference and pduction functions $$u(c)=\frac{c^{1-\frac{1}{\gamma}}}{1-\frac{1}{\gamma}} \quad \text{with} \quad u'(c)=c^{-\frac{1}{\gamma}},$$ where $\gamma$ denotes the intertemporal elasticity of substitution between consumption in different years. We have $$c_2=(\beta R^n)^\gamma c_1 \quad \text{and} \quad c_3=(\beta R^n)^\gamma c_2=(\beta R^n)^{2\gamma}c_1$$
$$W_1:=w^n+\frac{w^n}{R^n}+\frac{pen}{(R^n)^2}.$$
$$c_1=\Psi_1W_1 \quad \text{with} \quad \Psi_1=\frac{1}{p}\{1+\beta^\gamma(R^n)^{\gamma-1}+\beta^{2\gamma}(R^n)^{2(\gamma-1)}\}^{-1}.$$ The savings decisions are given by $$a_2=w^n-pc_1 \quad \text{and} \quad a_3=w^n+R^na_2-pc_2.$$ The production technology is assumed to be of Cobb-Douglas type $$f(k)=k^\alpha \quad \text{and} \quad f'(k)=\alpha k^{\alpha-1}.$$ Interest and wage rates are determined by $$r=\alpha k^{\alpha-1}-\delta \quad \text{and} \quad w=(1-\alpha)k^\alpha.$$

```python
##### Program 6.1 Computation of the long-run equilibrium

# Import package
import numpy as np
from scipy.optimize import root

# Parameters
γ    = 0.5  #Intertemporal elasticity of substitution between consumption (IES)
β    = 0.9  #discount factor
α    = 0.3  #capital share
δ    = 0.0  #depreciation
g    = np.array([0.12,0.12,0.0])  #per capita coefficient of public good provision
by   = 0.0  #debt proportion
κ    = 0.0  #pension wage ratio
n_p  = 0.2  #population growth rate
egam = 1 - 1/γ  #denominator of utility

# Taxes
τw   = 0.0  #labour income tax rate
τr   = 0.0  #capital income tax rate
τc   = 0.0  #consumption tax rate
tax  = 1    #choose which tax to balance the budget
#1 if consumption tax only, 2 if labour and capital income, 3 if labour income only, 4 if capitla income only


L    = (2 + n_p)/(1 + n_p)      #aggregate labour supply
τp   = κ/((2 + n_p)*(1 + n_p))  #payroll tax rate


def eqn(x):
    """
    This function defines the system where all markets should be clear and government budget balanced.
    Input:  K=x[0]: capital stock
            τ=x[1]: the consumption/labour-income/capital-income tax rates to balance the budget
    Output: equi[0]: Capital market surplus
            equi[1]: Government revenue-Government spending
    """
    global c1, c2, c3, a2, a3, τc, τw, τr, τp, A, L, Y, K, B, G, C, r, w
    K  = x[0]
    if   tax == 1:
        τc = x[1]  #government budget balance by consumption tax only
    elif tax == 2:
        τw = x[1]  #government budget balance by labour and capital income tax 
        τr = x[1]
    elif tax == 3:
        τw = x[1]  #government budget balance by labour income only
    else:
        τr = x[1]  #government budget balance by capital income only
    k   = K/L     #per capita capital stock
    # Factor price
    r   = α*k**(α - 1.0) - δ  #interest rate
    w   = (1.0 - α)*k**α      #wage rate
    wn  = w*(1.0-τw-τp)       #after tax wage rate
    Rn  = 1.0 + r*(1.0 - τr)  #after tax interest return
    p   = 1.0 + τc            #consumer price
    pen = κ*w                 #pension
    
    # Individual decisions
    ψ  = (1.0/p)*(1.0 + β**γ*Rn**(γ - 1.0) + β**(2.0*γ)*Rn**(2.0*(γ - 1.0)))**(-1.0)  #consumption fraction
    W  = wn + wn/Rn + pen/Rn**2   #total income
    c1 = ψ*W                      #consumption in period 1
    c2 = (β*Rn)**γ*c1             #consumption in period 2
    c3 = (β*Rn)**γ*c2             #consumption in period 3
    a2 = wn - p*c1                #savings to period 2
    a3 = wn + Rn*a2 - p*c2        #savings to period 3
    
    # Aggregation
    Y = K**α*L**(1.0-α)                            #output
    A = a2/(1.0+n_p)+a3/(1.0+n_p)**2.0             #aggregate savings
    C = c1 + c2/(1+n_p) + c3/(1+n_p)**2            #aggregate consumption
    B = by*Y                                       #public debt
    G = g[0] + g[1]/(1+n_p) + g[2]/(1.0+n_p)**2.0  #governemnt spending
    T = τc*C + τw*w*L + τr*r*A                     #government revenue (tax revenue)
    
    # Equilibrium conditions
    equi = np.zeros(2)            #initialize equi
    equi[0] = K + B - A           #capital market surplus
    equi[1] = T - G - (r-n_p)*B   #government surplus
    
    return equi

# Initial guess
x0 = np.array([0.5,0.5])

# Find root routine
res = root(eqn, x0) #find the k and τ that clears the capital market and balance the government budget

# Utility
U = (c1**egam + β*c2**egam + β**2*c3**egam)/egam  #expected utility for three periods

# Investment
I = (n_p+δ)*K

# Result
if res.success == True:
    print('-------------------------------------------------')
    print('c1'     ,'  |','c2'     ,'  |','c3'     ,'  |','Y '     ,'  |','w '     ,'  |','r '    ,'  |','U ')
    print("%.2f"%c1,  '|',"%.2f"%c2,  '|',"%.2f"%c3,  '|',"%.2f"%Y ,  '|',"%.2f"%w ,  '|',"%.2f"%r,  '|',"%.2f"%U)
    print('-------------------------------------------------')
    print('a2'     ,'  |','a3'     ,'  |','K '     ,'  |','C '     ,'  |','G '     ,'  |','I '    ,'  |')
    print("%.2f"%a2,  '|',"%.2f"%a3,  '|',"%.2f"%K ,  '|',"%.2f"%C ,  '|',"%.2f"%G ,  '|',"%.2f"%I,  '|')
    print('-------------------------------------------------')
    print('τw'     ,'  |','τr'     ,'  |','τc'     ,'  |','τp'     ,'  |','by'     ,'  |','κ '    ,'  |','np ')
    print("%.2f"%τw,  '|',"%.2f"%τr,  '|',"%.2f"%τc,  '|',"%.2f"%τp,  '|',"%.2f"%by,  '|',"%.2f"%κ,  '|',"%.2f"%n_p)
    print('-------------------------------------------------')
else:
    print('No equilibrium found !')
```

## 6.2 Transitional dynamics and welfare analysis

In this section we focus on the periods of so-called <i>transitional path</i> in which the economy successively adjusts from the initial steady state towards the new long-run equilibrium.

### 6.2.1 Computation of Transitional Dynamics
<b>Module extension</b> In order to account for transitional dynamics in our simulation model, we first need to make variables time dependent. 

<b>The main program</b> We first initialize aggregate labour supply, the payroll tax rate, and all other tax rates. In addition, we specify the change in policy or demographic parameters that should induce the transition path. In the case presented we change from a consumption to a purely labour-income tax system. Once everything is initialized, we compute the initial long-run equilibrium by finding the root of the equation system <i>eqns_Initial</i>. Having computed the initial equibiurm values, we compute the transition path and long-run equibrium using the subroutine <i>get_Transition</i>. 

<b>Prices and quantities along the transition path</b> Since the computationtime for the solution of the transitional dynamics equation system grows quadratically in the number of transition periods $TT$. We present an iterative solution method, often called the <i>Gauss-Diedel method</i>. This method is used in the following subroutines and programs:

1. We first initialize the capital stock along the transition at initial equilibrium values, $K_t=K_0$ for all $t \geq 1$, and assume that labour supply and tax rates have values as initialize in the main program.

2. With the capital stock $K_t$ and labour supply $L_t$ we can calculate the capital intensity $\frac{K_t}{L_t}$ and therefore derive factor prices $r_t$ and $w_t$ in the subroutine <i>factor_prices</i> for every year $t\geq 1$. We then use tax rates to determine net prices.

3. Next we use net prices in subroutine <i>decisions</i> to compute individual decisions of young cohorts.

4. We aggregate individual decisions to quantities $C_t, A_t, Y_t, I_t, G_t, B_t$, and $K_t$ in the subroutine <i>quantities</i>. The speed of adjustment is determined by the parameter $damp$. If this parameter is equal to 1, the capital stock immediately adjuests to completely clear the capital market. If $damp$ is small, adjustment takes place only gradually.

5. Knowing aggregated quantities, we determine budget-balancing tax and payroll rates for the government in the subroutine <i>government</i> for every period $t\geq 1$.

6. Finally we have to check whether the calculated quantities and prices define a general equilibrium. We do so by looking at the goods market equilibrium condition. We say that a goods market is in equilibrium if the absolute value of the relative difference between demand and supply is lower than some threshold level $tol$.

We cancel the iteration process after a maximum of $itermax$ iterations.

```python
## Function: factor price:
def factor_prices(i):
    """This function defines multidimensional factor price calculation for period i."""
    global k, K, r, w, wn, Rn, p, pen
    k[i]   = K[i]/L[i]                 #per capita capital stock
    r[i]   = α*k[i]**(α - 1.0)-δ       #interest rate
    w[i]   = (1.0 - α)*k[i]**α         #wage rate
    wn[i]  = w[i]*(1.0-τw[i] - τp[i])  #after tax wage rate
    Rn[i]  = 1.0 + r[i]*(1.0 - τr[i])  #after tax interest return
    p[i]   = 1.0 + τc[i]               #consumer price
    pen[i] = κ[i]*w[max(i-1,0)]        #pension
    return None
```

```python
## Calculates year at which age j agent is ij
def year(i,j,ij):
    """This function defines the year index"""
    year = i + ij -j  #year i age j period ij (j-1, j+1, j+2)
    if i == 0 or year <= 0:
        year = 0      #initial period
    if i == T or year >= T:
        year = T      #last period
    return year
```

```python
## Calculating individual decisions in a certain year

def decisions(i):
    """This function defines the decisions of consumption and savings in each period i."""
    global c, a
    
    # Calculate future and past years
    it1 = year(i, 1, 2)     #time index period i age 1 turns to age 2 (t+1)
    it2 = year(i, 1, 3)     #time index period i age 1 turns to age 3 (t+2)
    itm = year(i, 2, 1)     #time index period i age 2's previous period (t-1)
    
    # Individual decisions
    W       = wn[i] + wn[it1]/Rn[it1] + pen[it2]/(Rn[it1]*Rn[it2]) + v[i+1]  #expected wealth at period i
                                                                    #v is the compensating lump-sum transfer
    ψ       = (p[i]*(1.0 + β**γ*(p[it1]/(p[i]*Rn[it1]))**(1.0-γ) \
                     + β**(2*γ)*(p[it2]/(p[i]*Rn[it1]*Rn[it2]))**(1.0-γ)))**(-1.0) #consumption fraction
    c[0, i] = W*ψ                                                       #c_1 at period i (young)
    
    if i == 1 :  #period 1 tax reform starts
        W      = Rn[i]*a[1,0] + wn[i] + pen[it1]/Rn[it1] + v[1]         # W at period 1
        ψ      = (p[i]*(1.0 + β**γ*(p[it1]/(p[i]*Rn[it1]))**(1.0-γ)))**(-1.0)    # ψ at period 1
        c[1,i] = W*ψ                                                    #c_2 at period 1 (middle) 
        c[2,i] = (pen[i] + Rn[i]*a[2,itm] + v[0])/p[i]                  #c_3 at period 1 (old)
        a[1,i] = wn[itm] - p[itm]*c[0,itm]                              #a_2 at period 1 (middle)
    else:
        c[1,i] = (β*Rn[i]*p[itm]/p[i])**γ*c[0,itm]                      #c_2 at period t (middle) 
        c[2,i] = (β*Rn[i]*p[itm]/p[i])**γ*c[1,itm]                      #c_3 at period t (old)
        a[1,i] = wn[itm] + v[itm+1] - p[itm]*c[0,itm]                   #a_2 at period t (middle)
        
    if i == 2 :  #period 2 
        a[2,i] = wn[itm] + Rn[itm]*a[1,itm] + v[1] - p[itm]*c[1,itm]    #a_3 at period 2 (old)
    else:
        a[2,i] = wn[itm] + Rn[itm]*a[1,itm] - p[itm]*c[1,itm]           #a_3 at period 2 (old)
    return None
```

```python
## Calculating quantities in a certain year
def quantities(i):
    """This function defines the aggregated quantities of C, A, Y, I, G, B, K in period i."""
    global Y, C, G, A, B, K, I
    itm = year(i,2,1)                                                               #index for t-1
    
    Y[i] = K[i]**α*L[i]**(1.0 - α)                                                  #output
    C[i] = c[0,i] + c[1,i]/(1.0 + n_p[i]) + c[2,i]/((1.0 + n_p[i])*(1.0+n_p[itm]))    #consumption
    G[i] = g[0] + g[1]/(1.0 + n_p[i]) + g[2]/((1.0 + n_p[i])*(1.0+n_p[itm]))          #government spending
    A[i] = a[1,i]/(1.0 + n_p[i]) + a[2,i]/((1.0 + n_p[i])*(1.0+n_p[itm]))           #savings
    B[i] = by[itm]*Y[i]                                                             #debt
    
    if i == 0:
        I[i] = (n_p[i] + δ)*K[i]                                                    #investment at period 0
    
    if i > 0: #linear combination of the adjusted savings, debt
        K[i] = damp*(A[i] - B[i] - Ba[i]) + (1.0 - damp)*K[i]                       #capital stock
        I[i] = (1.0 + n_p[min(i+1, T)])*K[min(i + 1,T)] - (1.0-δ)*K[i]              #investment
    return None
```

```python
## Equation system to determine the initial equilibrium

def eqns_Initial(x):
    """
    This function defines the system where all markets should be cleared and government budget balanced at period 0.
    Input:  K=x[0]: capital stock
            τ=x[1]: consumption/labour-income/capital-income tax rates
    Output: eqns_Initial[0]: Capital market surplus
            eqns_Initial[1]: Government revenue-Government spending
    """
    global τc, τw, τr
    K[0] = x[0]
    if tax[0] == 1:
        τc[0] = x[1]   #government budget balance by consumption tax only
    elif tax[0] == 2:
        τw[0] = x[1]   #government budget balance by labour and capital income tax
        τr[0] = x[1]
    elif tax[0] == 3:
        τw[0] = x[1]   #government budget balance by labour income only
    else:
        τr[0] = x[1]   #government budget balance by capital income only
        
    factor_prices(0)   #include factor prices at period 0
    decisions(0)       #include decisions at period 0
    quantities(0)      #include quantities at period 0
    
    eqns_Initial    = np.zeros(2)           #initial set up, 2 entries
    eqns_Initial[0] = K[0] + B[0] - A[0]    #goods market clearing
    eqns_Initial[1] = τc[0]*C[0] + τw[0]*w[0]*L[0] + τr[0]*r[0]*A[0] - (r[0]-n_p[0])*B[0] - G[0]  #government budget
    
    return eqns_Initial
```

```python
## Calculate government parameters
def government(i):
    """This function defines the government parameters at period i."""
    global τc, τw, τr, τp
    
    # get budget balancing tax rate
    if tax[i] == 1:      #(government surplus-τcC)/C
        τc[i] = ((1.0 + r[i])*B[i] + G[i] - (τw[i]*w[i]*L[i] + τr[i]*r[i]*A[i] \
                + (1.0 + n_p[i])*B[min(i + 1,T)]))/C[i]    
    elif tax[i] == 2:    #[government surplus-τw(wL)-τr(rA)]/(wL+rA) assuming same labour and capital income tax
        τw[i] = ((1.0 + r[i])*B[i] + G[i] - (τc[i]*C[i] + \
                 (1.0 + n_p[i])*B[min(i + 1,T)]))/(w[i]*L[i] + r[i]*A[i])
        τr[i] = τw[i]    #same labour and capital income tax
    elif tax[i] == 3:    #[government surplus-τw(WL)]/(wL)
        τw[i] = ((1.0 + r[i])*B[i] + G[i] - (τc[i]*C[i] + τr[i]*r[i]*A[i] + \
                 (1.0 + n_p[i])*B[min(i + 1,T)]))/(w[i]*L[i])
    else:                #[government surplus-τr(rA)]/(rA)
        τr[i] = ((1.0 + r[i])*B[i] + G[i] - (τc[i]*C[i] + τw[i]*w[i]*L[i] + \
                 (1.0 + n_p[i])*B[min(i + 1,T)]))/(r[i]*A[i])
        
    # get budget balancing social security contribution
    τp[i] = (pen[i]/((2.0+n_p[i])*(1.0+n_p[i-1])))/w[i]    #payroll tax as evenly redistributed pensions
    
    return None
```

```python
## Compute household utility
def utility(i):
    """This function defines expected utility at each age for each period i in [0,T]."""
    global U
    if i == 0:
        it1 = 0
        it2 = 0
    else:
        it1 = min(i+1,T)
        it2 = min(i+2,T)
    
    # Oldest cohort
    U[2,i] = c[2,i]**egam/egam #1 period left
    
    # Middle cohort
    U[1,i] = c[1,i]**egam/egam + β*c[2,it1]**egam/egam #2 periods left
    
    # Youngest cohort
    U[0,i] = c[0,i]**egam/egam + β*c[1,it1]**egam/egam + β**2.0*c[2,it2]**egam/egam #all 3 periods
    
    return None
```

<b> Efficiency effects </b> In order to know whether welfare consequences are only due to intergenerational distribution or whether the reform has actual improved (or worsend) the resource allocation, we implement a Lump-Sum Redistribution Authority (LSRA). We allow this authority to pay lump-sum transfers to all generations. The LSRA thereby proceeds as follows: 

1. We calculate lump-sum transfers $v_{-1}$ and $v_0$ needed to be given to current generations to make then as well off as in the initial equilibrium. $$v_{-1}=[(\frac{U_{30}}{U_{31}})^{\frac{1}{1-1/\gamma}}-1]W_{31}\quad \text{and} \quad v_0=[(\frac{U_{20}}{U_{21}})^{\frac{1}{1-1/\gamma}}-1]W_{21}.$$
2. We calculate transfer payments $v_t, t=1,...T$ to future generations so as to make them all equally well off, i.e. they all experience the identical utility level $U^\ast$. The utility level $U^\ast$ is determined by the requirement that the present values of all transfer payments add up to 0 (LSRA does not make any profits). $$v_{-1}N_{-1}+v_0N_0+v_1N_1+\frac{v_2N_2}{R_2}+...+\frac{v_TN_TR_T}{\Pi_{k=2}^TR_k(r_T-n_{p,T})}=0.$$ Dividing $N_1$ and substituting the definition of $v_t$ yields $$\frac{v_-1}{(1+n_{p,0})(1+n_{p,1})}+\frac{v_0}{1+n_{p,1}}+\sum_{t=1}^{T-1}\big[\big(\frac{U^\ast}{U_t}\big)^{\frac{1}{1-1/\gamma}}-1\big]W_t\Pi_{s=2}^t\frac{1+n_{p,s}}{1+r_s}+\big[\big(\frac{U^\ast}{U_T})^{\frac{1}{1-1/\gamma}}-1\big]W_T\frac{1+r_T}{r_T-n_{p,T}}\Pi_{s=2}^T\frac{1+n_{p,s}}{1+r_s}=0.$$
3. We calculate the compensating lump-sum transfers for future cohorts given $U^\ast$: $$v_t=\big[\big(\frac{U^\ast}{U_t}\big)^{\frac{1}{1-1/\gamma}}-1\big]W_t \quad \text{for} \quad t=1,...,T,$$ where $\Delta^\ast=\big(\frac{U^\ast}{U_t}\big)^{\frac{1}{1-1/\gamma}}-1$ represents the welfare change.
Debt (or assets) of the agency per capita of workers in period 2 are given by $$B_2^a=\frac{1}{1+n_{p,2}}\big[\frac{v_{-1}}{(1+n_{p,0})(1+n_{p,1})}+\frac{v_0}{1+n_{p,1}}+v_1\big].$$ The periodic budge constriant is: $$(1+n_{p,t+1})B_{t+1}^a=(1+r_t)B_t^a+v_t.$$ The capital market equilibrium condition with LSRA becomes: $$A_t=K_t+B_t+B_t^a \quad \text{for} \quad t=2,...,T.$$

```python
## Calculating lsra transfers

def lsra():
    """This function calculates social welfare. True to include, False to exclude."""
    global Wl, v, Ba, ustar
    
    #calculate utility for each generation
    for i in range(1,T+1):
        utility(i)    #get utility for each period i=0,1,2,...T
        
    # transfers to old generations
    Wl    = np.zeros(T+2)      #initialize T+2 values of wealth (resources of older generations)
    #v     = np.zeros(T+2)      #initialize T+2 values of compensating lump-sum transfer
    Wl[0] = Rn[1]*a[2,1] + pen[1] + v[0]                                #wealth at period 0
    Wl[1] = Rn[1]*a[1,1] + wn[1] + pen[2]/Rn[2] +v[1]                   #wealth at period 1
    #v: required transfer that bring older cohorts in the reform year from their current utility level 
        #back to their initial equilibrium utilities
    v[0]  = v[0] + Wl[0]*((U[2,0]/U[2,1])**(1.0/egam)-1.0)  #v-1        #from U_{3,1} to U_{3,0} (old generation)
    v[1]  = v[1] + Wl[1]*((U[1,0]/U[1,1])**(1.0/egam)-1.0)  #v0         #from U_{2,1} to U_{2,0} (middle generation)
    Ba[2] = v[0]/((1.0 + n_p[0])*(1.0 + n_p[1])) + v[1]/(1.0 + n_p[1])  #debt or assets of the agency per capita of workers
    
    #Calculate the following separated terms of the equation to solve for U^*.
    # long run equilibrium (last period) 
    Wl[T+1] = wn[T] + wn[T]/Rn[T] + pen[T]/Rn[T]**2.0 + v[T+1]          #old generation wealth at period T+1
    Pvv     = v[T+1]*(1.0+r[T])/(r[T]-n_p[T])                           #v(1+r)/(r-n_p) in period T part
    sum1    = Wl[T+1]*(1.0+r[T])/(r[T]-n_p[T])                          #wl(1+r)/(r-n_p) in period T part
    sum2    = Wl[T+1]*(U[0,T]*egam)**(-1.0/egam)*(1.0+r[T])/(r[T]-n_p[T])  #normalized Wl/Ut in period T part
    
    # transition path (backward induction)
    for i in range(T-1,0,-1): # from T to 1.
        it1     = year(i, 1, 2)   #year index as t+1
        it2     = year(i, 1, 3)   #year index as t+2
        Wl[i+1] = wn[i] + wn[it1]/Rn[it1] + pen[it2]/(Rn[it1]*Rn[it2]) + v[i+1] #wealth at period i
        Pvv     = Pvv*(1.0 + n_p[it1])/(1.0 + r[it1]) + v[i+1]                  #v(1+np)/(1+r) part
        sum1    = sum1*(1.0 + n_p[it1])/(1.0 + r[it1]) + Wl[i+1]                #Wl(1+np)/(1+r) part
        sum2    = sum2*(1.0 + n_p[it1])/(1.0 + r[it1]) + Wl[i+1]*(U[0,i]*egam)**(-1.0/egam)  # normalized Wl/Ut part
    
    # calculate ustar for future generations 
    ustar = ((sum1 - Ba[2] - Pvv)/sum2)**egam/egam  #the identical utility level for future generations 
    
    # calculate transfers to future generations and debt of LSRA
    for i in range(1,T+1):
        v[i+1] = v[i+1] + Wl[i+1]*((ustar/U[0,i])**(1.0/egam) - 1.0)  #lump-sum transfers for future cohorts
        if i == 2:
            Ba[2] = (Ba[2] + v[2])/(1.0 + n_p[2])                 #B_{t+1}=(above Ba+v)/(1+np) equation on page 243
        if i > 2:
            Ba[i] = ((1.0 + r[i-1])*Ba[i-1] + v[i])/(1.0 + n_p[i])#B_{t+1}=((1+r)Ba+v)/(1+np) same equation on page 243
            
    return None
```

```python
## Solves for transition path using Gauss-Seidel

def get_Transition():
    """This function calculates each period economy and checks for market clearing conditions."""
    
    # initialize values from initial equilibrium
    K[:] = K[0]  #initial capital
    nmarket = 0  #iteration index
    
    for ite in range(1,itermax+1):
        
        # get prices, decisions and quantities for each period i
        for i in range(1,T+1):
            factor_prices(i)
            
        for i in range(1,T+1):
            decisions(i)
            
        if lsra_on: #true or false to include/exclude welfare analysis
            lsra()
        
        for i in range(1,T+1):
            quantities(i)
        
        for i in range(1,T+1):
            government(i)
        
        # check for the number of markets in equilibrium and whether cleared
        nmarket = 0
        for i in range(1,T+1):
            if abs(Y[i] - C[i] - I[i] - G[i])/Y[i] < tol : #if yes, consider as market cleared
                nmarket = nmarket + 1
        
        print('Iteration: ', ite, 'Markets: ', nmarket, 'Diff', np.max(np.absolute(Y - C - I - G)/Y))
        
        if nmarket == T:
            break #exist loop
    
    if ite == itermax:
        print('!!! No equilibrium found !!!')
    
    return None
```

```python
##### Program 6.2 The OLG model with transitional dynamics

# Import package
import numpy as np

# Parameters
T = 25                #transition periods
γ = 0.5               #IES or risk aversition
egam = 1 - 1/γ        #denominator of CRRA utility function for easier use
β = 0.9               #discount factor
α = 0.3               #capital share
δ = 0.0               #depreciation
tol = 1e-5            #tolerance
damp = 0.25           #converging speed
itermax = 1000        #maximum iteration
g = np.array([0.12, 0.12, 0.0])   #government parameter g1, g2, g3

# Time-variation variables
by = np.zeros(T+1)         #initial debt proportion
κ = np.zeros(T+1)          #initial pension wage ratio
n_p = np.zeros(T+1) + 0.2  #initial growth rate
tax = np.ones(T+1)         #initial tax scheme
lsra_on = False            #True: open for welfare analysis; False: close for welfare analysis

# Model variables all set to be zero
w   = np.zeros(T+1); r = np.zeros(T+1); wn = np.zeros(T+1); Rn = np.zeros(T+1); p = np.zeros(T+1); τp = np.zeros(T+1)
pen = np.zeros(T+1); K = np.zeros(T+1); L  = np.zeros(T+1); Y  = np.zeros(T+1); A = np.zeros(T+1); C  = np.zeros(T+1)
I   = np.zeros(T+1); B = np.zeros(T+1); G  = np.zeros(T+1); Ba = np.zeros(T+1); k = np.zeros(T+1)
a   = np.zeros((3,(T+1))); c = np.zeros((3,(T+1))); U = np.zeros((3,(T+1))); v = np.zeros(T+2)

# Set Reform values (uncomment to get the corresponding table)
tax[1:] = 2                                #table 6.2
# tax[1:] = 4                                #table 6.3
# tax[1:] = 3                                #table 6.4
# tax[1:] = 3; by[1:] = -0.058857            #table 6.5
# κ[1:] = 0.5                                #table 6.6
# by[1:] = 0.0986                            #table 6.7
# n_p[1:] = 0.0                              #table 6.8
# n_p[1:] = 0.0; κ[:] = 0.5; g[:] = 0.0      #table 6.9

# Get labour supply and pension payments
L = (2.0 + n_p)/(1.0 + n_p)       #aggregate labour supply
τp = κ/((2.0 + n_p)*(1.0 + n_p))  #payroll tax rate

# Initialize tax rates
τw = np.zeros(T+1); τr= np.zeros(T+1); τc = np.zeros(T+1)

# Initialize assets, LSRA payments and debt holdings
a  = np.zeros((3,(T+1))); v = np.zeros(T+2); Ba = np.zeros(T+1);

# Initial Guess
x0 = np.zeros(2) + 0.7

# Solve the steady state equation system
res = root(eqns_Initial, x0)

# Check whether the solution is valid
if res.success == False:
    print('!!! No equilibrium found !!!')
    
get_Transition()
print('done with no lsra')
for i in range(T+1):
    utility(i)

U_no_lsra = U.copy()
```
