Here we plot the graph of function $f(x)=x^2$ and $f(x)=\sqrt{x}$ with appropriate labels and legend.

```python
##### Program 1.17 Plotting graphs with the matplotlib.pyplot module

## Import module
import matplotlib.pyplot as plt
import numpy as np

### Plot function f(x)=x^2
## Initialize the x values
# Create 101 points of x in the interval [0,1] with equally spaced.
x = np.linspace(0,1,101)

# Calculate the square of x and store them in vector y_square
# The vector y is the element-wise square of vector x.
y_square = x**2

# Here we have two arrays to plot. Uncomment to see the arrays
#print('Horizontal axis: ',x)
#print('Vertical axis: ',y_square)

# Using matplotlib.pyplot module to plot these two arrays, that is the graph of f(x)=x^2.
plt.figure() # create a new figure
plt.clf() # clear any existing plots ('clf' = 'clear figure')
plt.plot(x, y_square) # generate the plot
plt.show() # view the plot

### Plot function f(x)=x^(1/2)
## Calculate the square roots of x and store them in vector y_squareroot
y_squareroot = x**(1/2)

# Here we have another two arrays to plot. Uncomment to see the arrays
#print('Horizontal axis: ',x)
#print('Vertical axis: ',y_squareroot)

# Using matplotlib.pyplot module to plot these two arrays, that is the graph of f(x)=x^(1/2).
plt.figure() 
plt.clf() 
plt.plot(x, y_squareroot) 
plt.show() 

### Plot the square root and cubic root into one graph.
## Calculate the square roots of x and store them in vector y_cubicroot
y_cubicroot = x**(1/3)

# Plot these two arrays
plt.figure()
plt.clf()
plt.plot(x, y_squareroot, label='square root') # plot x versus x^(1/2) and label it as 'square root'
plt.plot(x, y_cubicroot,label='cubic root') # plot x versus x^(1/3) and label it as 'cubic root'
plt.legend(loc='best') # display legend and using the default best display location
plt.show()

## Repeat plotting square root and cubic root with different line colors and styles.
plt.figure()
plt.clf()
plt.title('Roots') # Set the title of this figure as 'Roots'.
plt.plot(x, y_squareroot, color='r', linestyle='-.', linewidth=1,label='square root') 
# Plot square root function with red dash-dot line and set line thickness=1.5.
plt.plot(x, y_cubicroot, color='b', linestyle='--', linewidth=2,label='cubic root')
# Plot cubic root function with bule dash line and set line thickness=1.5.
plt.legend(loc='best') # display legend and using the default best display location
plt.show()

## Instead, we can plot these two functions with markers.
plt.figure()
plt.clf()
plt.title('Roots') # Set the title of this figure as 'Roots'.
plt.plot(x, y_squareroot,'r.', markersize=3,label='square root') 
# make the plot red, and mark each point with a dot
plt.plot(x, y_cubicroot, 'b+',markersize=3,label='cubic root')
# make the plot bule, and mark each point with a +
plt.legend(loc='best') # display legend and using the default best display location
plt.show()

```
